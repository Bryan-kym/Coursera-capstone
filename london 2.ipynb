Now we clean the data and leave the necessary informsstion which we will use to come up with the solution for our problem

```python
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

## METHODOLOGY <a name="methodology"></a>

We will put our efforts into finding areas in London that have a low density in restaurants and specifically each category of restaurant. 
since we want to identify each type of restaurant, we will first cluster them into groups based on theiur cuisines and use the folium maps to showcase the restaurants and identify the areas that are least concentrated with a specific category of restaurants one at a time until we have showcased all the categories.

We have already collected the locatioon data, the retstaurants themselves and the category of each. 

## ANALYSIS <a name="analysis"></a>

```python
# one hot encoding
london_onehot = pd.get_dummies(dataframe_filtered[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
london_onehot['name'] = dataframe_filtered['name'] 

# move neighborhood column to the first column
fixed_columns = [london_onehot.columns[-1]] + list(london_onehot.columns[:-1])
london_onehot = london_onehot[fixed_columns]

london_onehot.head()
```

```python
london_grouped = london_onehot.groupby('name').mean().reset_index()
london_grouped
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
```

Lets cluster the retsaurants based on their categories.

We will use 16 clusters

```python
kclusters = 16

london_grouped_clustering = london_grouped.drop('name', 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(london_grouped_clustering)

kmeans.labels_[0:16] 
```

```python
london_grouped.insert(0, 'Cluster Labels_1', kmeans.labels_)

london_merged = dataframe

london_merged = london_merged.join(london_grouped.set_index('name'), on='name')

london_merged.head()
```
